```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Input data
time = np.arange(1, 61)
df = pd.read_csv('data csv/15.csv')
data = df['percobaan_5']
```

```python
# Get the dynamic shape based on the size of the data
heatmap_shape = (1, len(data))
# Reshape data for heatmap
heatmap_data = np.array(data).reshape(*heatmap_shape)
# Create heatmap
plt.figure(figsize=(12, 8))
# heatmap themes = viridis, magma
sns.heatmap(heatmap_data, annot=False, fmt=".5f", cmap="plasma", xticklabels=10, yticklabels=10, cbar=False, vmin=0.7096, vmax=2.94)
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.xticks([])
plt.yticks([])
plt.axis('off')
# plt.savefig('1D_Heatmap.png', bbox_inches='tight', pad_inches=0, transparent=True)
plt.show()  
```
